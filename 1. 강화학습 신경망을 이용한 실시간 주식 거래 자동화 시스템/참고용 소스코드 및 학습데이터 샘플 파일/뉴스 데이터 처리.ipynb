```python
import pandas as pd
import numpy as np
from collections import Counter
import re
```

```python
aaa = pd.read_csv('./newsdata/삼성전자 뉴스 데이터.csv', encoding='utf-8')
abc = aaa.copy()
abc
```

```python
word_list = []

for i in range(0, abc.shape[0]):
    content_str = str(abc.iloc[i]['내용'])
    for word in content_str.split():
        word_list.append(word)
```

```python
word_list_count = Counter(word_list)
word_list_count
```

```python
zzz = pd.DataFrame([dict(word_list_count)])
zzz
```

```python
word_df = pd.DataFrame({
      '단어':zzz.columns
    , '개수':zzz.values[0]
})
word_df
```

```python
word_df.to_csv('./삼성 키워드.csv', encoding='utf-8', index=False)
```

```python
load = pd.read_csv('./삼성 키워드.csv', encoding='utf-8')
load
```

```python
filteringWord = []
for i in range(0, load.shape[0]):
    resultText = re.sub('[0-9a-zA-Z-=+,#/\?:^.@%$&_△*\"※~ㆍ!』‘|\(\)\[\]`\'…》\”\“\’·\s]', '', str(load['단어'][i]))
    filteringWord.append(resultText)
```

```python
dadada = pd.DataFrame({
    '단어':filteringWord
})
```

```python
dadada.to_csv('./삼성 필터 처리된 키워드.csv', encoding='utf-8', index=False)
```
