```python
import numpy as np
import pandas as pd
import talib
```

```python
data = pd.read_csv('./Stock_Trading_System/32bit/hapdaystockdata/루닛 일 단위 주식 데이터.csv', encoding='euc-kr')
dataCp = data.copy()
# dataCp.drop(['거래대금','누적체결매도수량','누적체결매수수량'], axis=1, inplace=True)
dataCp
```

```python
# SMA5 -> 5일동안 이동평균
# upper, middle, lower -> 볼린저 밴드
# RSI -> 상대강도지수
# MACD -> 이동 평균 수렴 발산
# AROON -> 기간의 최고점과 최저점을 분석하여 추세의 강도와 방향을 측정하는 추세 추종 지표
upper, middle, lower = talib.BBANDS(dataCp['종가'], timeperiod=20)
macd, macdsignal, macdhist = talib.MACD(dataCp['종가'], fastperiod=12, slowperiod=26, signalperiod=9)
ar_up, ar_dn = talib.AROON(dataCp['고가'], dataCp['저가'], timeperiod=14)
df = pd.DataFrame({
    '종목코드':dataCp['종목코드'],
    '종목명':dataCp['종목명'],
    '날짜':dataCp['날짜'],
    'SMA5':talib.SMA(dataCp['종가'], timeperiod=5),
    'SMA20':talib.SMA(dataCp['종가'], timeperiod=10),
    'UPPER':upper,
    'MAVG':middle,
    'LOWER':lower,
    'RSI':talib.RSI(dataCp['종가'], timeperiod=14),
    'MACD':macd,
    'AROONUP':ar_up,
    'AROONDN':ar_dn
})
df[33:]
```
