```python
chtl = []
for i in range(dayDf.shape[0]):
    changeTime = dt.datetime(int(str(dayDf['날짜'][i])[:4]), int(str(dayDf['날짜'][i])[4:6]), int(str(dayDf['날짜'][i])[6:8]))
    chtl.append(changeTime.strftime('%Y-%m-%d %H:%M'))
dayDf['날짜'] = chtl
```
